# 4. On average which region has the highest

# a. GDP per capita (current US$)


```python
gdp=country.groupby(['Region']).mean().sort_values('GDP per capita (current US$)', ascending=False)
gdp.loc[:,['GDP per capita (current US$)']]
```

# b. International trade: Imports (million US$)


```python
imp=country.groupby(['Region']).mean().sort_values('International trade: Imports (million US$)', ascending=False)
imp.loc[:,['International trade: Imports (million US$)']]
```

# 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
cons=country.groupby(['Region']).std().sort_values('GDP per capita (current US$)')
cons.loc[:,['GDP per capita (current US$)']]
```
