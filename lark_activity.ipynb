```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cust_data=pd.read_excel('Telco-Customer-Churn.xlsx')
```

```python
cust_data
```

# 1. Compare churn count with respect to gender.

```python
print("Gender-wise Churn Count \n",cust_data[['gender','Churn']].value_counts())
sns.countplot(data=cust_data,x='gender',hue='Churn')
plt.title('Churn Count-Male Vs Female')
```

# 2. Find out how many female senior citizens are there in the dataset

```python
cust_data[['gender','SeniorCitizen']].value_counts()
```

# 3. Compare 'tenure' with 'Total Charges'


```python
plt.figure(figsize=(8,6))
plt.scatter(cust_data['tenure'],cust_data['TotalCharges'],s=8)
plt.xlabel('Tenure', fontsize =12)
plt.ylabel('Total Charges', fontsize =12)
plt.title('Tenure Vs Total Charges', fontweight='bold',fontsize=16)
plt.show()
```

# 4. Find out which contract is preferred by the senior citizen.

```python
senior=cust_data[cust_data['SeniorCitizen']==1]
```

```python
print('Contract preference of Senior Citizens \n', senior['Contract'].value_counts())
```

```python
plt.figure(figsize=(6,6))
plt.grid()
sns.countplot(data=senior,x='Contract')
plt.title('Contract Preference of Senior Citizens')
```

# 5. Comment on your finds on Payment Method?

```python
cust_data['PaymentMethod'].value_counts()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=cust_data,x='PaymentMethod')
plt.grid()
plt.xticks(rotation=15)
plt.title('Preferebility of different payment methods')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.countplot(data=cust_data,x='PaymentMethod',hue='gender')
plt.grid()
plt.xticks(rotation=15)
plt.title('Gender-wise payment method preferability')
plt.show()
```
