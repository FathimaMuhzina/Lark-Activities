# Preprocessing

```python
#Checking for null values
df.isna().sum()
```

```python
#Checking for duplicate rows
df[df.duplicated(subset=None, keep= 'first')==True]
```

```python
#Checking number of unique values in each column
df.nunique()
```

```python
#Checking datatypes
df.dtypes
```

```python
#Labelling all binary columns(yes/no) with 1/0
for col in df.columns:
    if((df[col].dtype=='object') and df[col].isin(['yes','no']).any()):
        df[col]=df[col].map({'yes':1, 'no':0})
df.head()
```

```python
#One-hot encoding for all categorical variables
df=pd.get_dummies(df)
df       
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max=MinMaxScaler(feature_range=(0,1))
df_scaled = pd.DataFrame(min_max.fit_transform(df),columns = df.columns)
```

```python
#Separating X and y
X=df_scaled.drop(['passed'],axis=1)
y=df_scaled['passed']
```

```python
#Splitting the dataset into training and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size=0.25)
X_train.shape
```

# Training the model with different algorithms

```python
#Creating a function for training the data and to return training time.
from time import time


table={'Training Time':[],'Prediction Time(Test)':[],'F1 Score(train)':[],'F1 Score(test)':[]}
comp_tab=pd.DataFrame(table)

def train_classifier(clf, X_train, y_train):
    start = time()
    clf.fit(X_train, y_train)
    end = time()
    train_time=round(end-start,4)
    return train_time
```

```python
#Creating a funcion for prediction and to return prediction time and corresponding F1 Score.
from sklearn.metrics import f1_score
def predict_labels(clf, features, target):
    start = time()
    y_pred = clf.predict(features)
    end = time()
    f1=f1_score(target.values, y_pred)
    pred_time=round(end-start,4)
    return pred_time,f1
```


    We are using these 6 common algorithms.
    1. Logistic Regression
    2. Support Vector Machine (SVM)
    3. K  Nearest Neighbour (KNN)
    4. Decision Trees
    5. Random Forest
    6. Gradient Boosting


```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
LR = LogisticRegression()
SVM = SVC()
KNN = KNeighborsClassifier()
DT = DecisionTreeClassifier()
RF = RandomForestClassifier()
GB = GradientBoostingClassifier()
for clf in [LR, SVM, KNN, DT, RF, GB]:
    tt=train_classifier(clf, X_train, y_train)
    pt_train,f1_train=predict_labels(clf, X_train, y_train)
    pt_test,f1_test=predict_labels(clf, X_test, y_test)
    comp_tab.loc[len(comp_tab.index)] = [tt,pt_test,f1_train,f1_test]
    
comp_tab.insert(0, 'Algorithm', ['LR', 'SVM', 'KNN', 'DT', 'RF', 'GB']) 
comp_tab
```

    Inference
    In terms of training and prediction time, Decision Tree is best. 
    In terms of F1 scores, Random Forest is best.
    Since we have to choose one with least amount of computational cost, we have to consider both time and accuracy. 
    Therefore,  we may choose Logistic Regression.
