2. How many patients in the data are having the blood pressure level above 120?

4.Find the minimum and maximum value of insulin level in the data.
5. Create a histogram of "Age".

7. Replace the min value of "BMI" with mean value of BMI

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
Diabetes_Model= pd.read_csv(r'C:\Users\user\OneDrive\Desktop\Data Science\Week#3\diabetes_model.csv')
```

# Display the Glucose level and the Age for the samples having the index ranging from 45 to 60.

```python
Diabetes_Model.loc[45:60,["Glucose","Age"]]
```

# How many patients in the data are having the blood pressure level above 120?

```python
Df2= Diabetes_Model[Diabetes_Model["BloodPressure"] > 120]
print("No of patients having Blood Pressure >120 is " ,Df2.shape[0])
print("Details of the patient are as per below table\n",)
Df2
```

# Select the rows the age is between 24 and 29 (inclusive)

```python
Df3=Diabetes_Model.loc[(Diabetes_Model['Age']>=24) & (Diabetes_Model['Age']<=29)]
print("Number of rows having age from 24 to 29 is  ", Df3.shape[0])
print("Row details are as below ",)
Df3
```

# Find the minimum and maximum value of insulin level in the data.

```python
Dia_Model=Diabetes_Model["Insulin"]
Dia_Model.describe().loc[["min","max"]]
```

# Create a histogram of "Age".

```python
plt.figure()
plt.hist(Diabetes_Model['Age'],color="Green",rwidth=0.9)
plt.title('Histogram of Age',fontsize=20)
plt.xlabel("Age",fontsize=15,color="Brown")
plt.ylabel("Frequency",fontsize=15,color="Brown")
```

# Create a scatter plot of "BMI" vs "Glucose".

```python
plt.figure(figsize=(10,10))
plt.scatter(Diabetes_Model["BMI"],Diabetes_Model["Glucose"],s=20,c="blue",marker="*")
plt.title("Scatter Plot of BMI vs Glucose ",fontsize=16)
plt.xlabel("BMI",fontsize=12)
plt.ylabel("Glucose",fontsize=12)
```

# Replace the min value of "BMI" with mean value of BMI

```python
print("Mean value of BMI\n",Diabetes_Model["BMI"].mean())
print("Min value of BMI\n",Diabetes_Model["BMI"].min())
```

# Draw a grouped boxplot with "Outcome" on x-axis and "Glucose" on the y-axis such that we see distributions of Glucose separately for each outcome.

```python
plt.figure(figsize=(10,10))
sns.boxplot(x="Outcome",y="Glucose",data=Diabetes_Model)
plt.title("Boxplot of outcome vs glucose ")
```

### 5.

**-** Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

**-** Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

**-** Combine the arrays first and second vertically, the final array size should be 10 x 5

# Create a numpy array (first) with 5 x 5 size and all elements are 0 in that.

```python

import numpy as np
arr1=np.zeros((5,5),dtype=int)
print(arr1)
```

# Create a numpy array (second) with 5 x 5 and all elements are 1 in that.

```python
arr2=np.ones((5,5),dtype=int)
print(arr2)
```

# Combine the arrays first and second vertically, the final array size should be 10 x 5

```python
arr3=np.concatenate((arr1,arr2),axis=0)
print(arr3)
```
