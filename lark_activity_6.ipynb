```python
import numpy as np
import pandas as pd
country=pd.read_excel(r'Dataset_w6.xlsx')
```

# 1. Find out how many countries are there in the dataset.


```python
country['country'].count()
```

# 2. Examine the data by using visualizations.

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(8,6))
plt.scatter(country['GDP per capita (current US$)'],country['Agricultural production index '],s=8)
plt.xlabel('GDP per capita (current US$)', fontsize =12)
plt.ylabel('Agricultural production index ', fontsize =12)
plt.title('GDP per capita Vs Agricultural Index', fontweight='bold',fontsize=16)
plt.show()
```

```python
most_pop=country.nlargest(5,['Population in thousands (2017)'])
plt.pie(data=most_pop,x='Population in thousands (2017)',labels=most_pop['country'],autopct='%1.1f%%')
plt.title('Top 5 Most Populated Countries ')
```

```python
largest=country.nlargest(5,['Surface area (km2)'])
plt.pie(data=largest,x='Surface area (km2)',labels=largest['country'],autopct='%1.1f%%')
plt.title('Top 5 Largest Countries ')
```

# 3. Find out how many countries with respect to the region are available in the dataset


```python
pd.value_counts(country['Region'])
```

# 4. On average which region has the highest

# a. GDP per capita (current US$)


```python
gdp=country.groupby(['Region']).mean().sort_values('GDP per capita (current US$)', ascending=False)
gdp.loc[:,['GDP per capita (current US$)']]
```

# b. International trade: Imports (million US$)


```python
imp=country.groupby(['Region']).mean().sort_values('International trade: Imports (million US$)', ascending=False)
imp.loc[:,['International trade: Imports (million US$)']]
```

# 5. Find out which region has more consistent with respect to GDP per capita (current US$)

```python
cons=country.groupby(['Region']).std().sort_values('GDP per capita (current US$)')
cons.loc[:,['GDP per capita (current US$)']]
```
